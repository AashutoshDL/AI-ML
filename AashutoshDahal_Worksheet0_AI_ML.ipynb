```python
#takes the type and the value
def convertLen(x,type):
  if (type=="meters"):
    result=x*3.28
  elif(type=="feet"):
    result=x/3.28
  return result

#takes weigth and the type accodingly
def convertWeght(x,type):
  if (type=="kilograms"):
    result=x*2.20
  elif(type=="pounds"):
    result=x/2.20
  return result

#takes volume and the type
def convertVol(x,type):
  if (type=="liters"):
    result=x*3.78
  elif(type=="gallons"):
    result=x/3.78
  return result
```

```python
conversion=input("Enter the type of conversion: length, weight, volume")
a=input("Enter the value you want to covert: ")
try:
    a=int(a)
except:
    print("Sorry invalid input")
type=input("Enter the type of value your entered: meters,feet,kilograms,pounds,liters,gallons")
match conversion:
  case "length":
    print(convertLen(a,type))
  case "weight":
    print(convertWeght(a,type))
  case "volume":
    print(convertVol(a,type))
```

Question no 2

```python
def sum(x):
  # Takes a list as input and returns the sum of its elements
  result = 0
  for num in x:
    result += num
  return result

def average(x):
  # Takes a list as input and returns the average of its elements
  result = 0
  for num in x:
    result += num
  result = result / len(x)
  return result

def maximum(x):
  # Takes a list as input and returns the maximum value from the list
  return max(x)

def minimum(x):
  # Takes a list as input and returns the minimum value from the list
  return min(x)
```

```python
a=input("Enter a list that is seperated by spaces")
try:
  list=[int(x) for x in a if x.isnumeric()]
except:
  print("Please enter valid numbers")
operation=input("Enter the operation: sum, average,maximum,minimum")
match operation:
  case "sum":
    print(sum(list))
  case "average":
    print(average(list))
  case "maximum":
    print(maximum(list))
  case "minimum":
    print(minimum(list))
```

Question no 12

```python
from collections.abc import Iterable

def sum_nested(lst):
  total = 0
  for x in lst:
    if isinstance(x, Iterable) and not isinstance(x, str):
      for y in x:
        if isinstance(y, Iterable) and not isinstance(y, str):
          for z in y:
            total += z
        else:
          total += y
    else:
      total += x
  return total

print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
def removeElement(lst,elem):
  for x in lst:
    for y in x:
      if(y==elem):
        x.remove(y)
  return lst

print(removeElement([[1, 2], [3, 2], [4, 5]],2))
```

Question no 14

```python
def find_max(lst):
  max_val = 0
  for x in lst:
    if type(x) == list:
      for y in x:
        if type(y) == list:
          for z in y:
            if z > max_val:
              max_val = z
        else:
          if y > max_val:
            max_val = y
    else:
      if x > max_val:
        max_val = x
  return max_val

print(find_max([[1, 2], [3, [4, 5]], 6]))
```

Question no 16

```python
def deepFlatten(lists):
  newList=[]
  for x in lists:
    if(type(x)==list):
      for y in x:
        if(type(y)==list):
          for z in y:
            newList.append(z)
        else:
            newList.append(y)
    else:
      newList.append(x)
  return newList

print(deepFlatten([[1, 2], [3, [4, 5]], 6]))
```
